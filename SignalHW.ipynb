```python
# DC Signal
import numpy as np
import matplotlib.pyplot as plt 
time = np.linspace(0, 2, 100)
voltage = 5 + time*0    #For 5V that is not changing because of DC voltage

plt.title('DC Signal')  #to give title in the plot
plt.xlabel('Time')      #Defining plot x axis 
plt.ylabel('Voltage')   #Defining plot Y axis 

plt.plot(time, voltage) #Ploting 
plt.show()              #Show the plot

```

```python
# Exponential Signal
import numpy as np
import matplotlib.pyplot as plt
time = np.linspace(0, 100, 50)
voltage = np.exp(time)              #this is for exponential voltage

plt.title('Exponential Signal')     #to give title in the plot
plt.xlabel('Time')                  #Defining plot x axis 
plt.ylabel('Voltage')               #Defining plot Y axis 

plt.plot(time,voltage,color="red")  #Ploting 
plt.show()                          #Show the plot
```

```python
# Square Signal
import numpy as np
from scipy import signal              #Import signal from scipy
import matplotlib.pyplot as plt
time = np.linspace(0, 1000, 1000)
voltage = signal.square(4*np.pi*time) #square voltage 

plt.title('Square Signal')            #to give title in the plot
plt.xlabel('Time')                    #Defining plot x axis 
plt.ylabel('Voltage')                 #Defining plot Y axis 

plt.plot(time,voltage,color="yellow") #Ploting 
plt.show()                            #Show the plot
```

```python
# Triangular Signal
import numpy as np
from scipy import signal                 #Import signal from scipy
import matplotlib.pyplot as plt

time = np.linspace(0, 1000, 1000)
voltage = signal.sawtooth(8*np.pi*time)  #to make the signal triangular

plt.title('Triangular Signal')           #to give title in the plot
plt.xlabel('Time')                       #Defining plot x axis 
plt.ylabel('Voltage')                    #Defining plot Y axis 

plt.plot(time,voltage,color="violet")    #Ploting 
plt.show()                               #Show the plot
```

```python
# Unit Impulse Signal
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt

time = np.linspace(0, 10, 100)
voltage = signal.unit_impulse(100)     #For unit pulse signal 

plt.title('unit impulse Signal')       #to give title in the plot
plt.xlabel('Time')                     #Defining plot x axis 
plt.ylabel('Voltage')                  #Defining plot Y axis 

plt.plot(time,voltage,color="orange")  #Ploting 
plt.show()                             #Show the plot
```

```python
# Unit step function : Continuous
import matplotlib.pyplot as plt
import numpy as np

time = np.linspace(0,2,100)
voltage = 1 + time*0                           # For unit step function

plt.title("Unit Step Function : Continuous")   # Title
plt.xlabel("Time")                             # Labeling
plt.ylabel("Voltage")                          # Labeling

plt.yticks([1])                                # This one removes ticks in the y axis other than 1
plt.plot(time, voltage)                        # Ploting
```

```python
#Unit Step Function : Discrete
import matplotlib.pyplot as plt
import numpy as np

time = np.linspace(0,20,10)
voltage = 1 + time*0                       # For unit step function: Discrete

plt.title("Unit Step Function : Discrete") # Title
plt.xlabel("Time")                         # Labeling 
plt.ylabel("Voltage")                      # Labeling

plt.yticks([1])                            # This one removes ticks in the y axis other than 1

plt.plot(time, voltage, "ro")              # Ploting the graph with dot 
plt.show()                                  
```
